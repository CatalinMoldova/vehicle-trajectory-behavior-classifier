# Exploratory Trajectory Analysis

EDA on the public synthetic sample dataset. Shared logic is imported from `vehicle_behavior`.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from vehicle_behavior.data import extract_vehicle_labels, load_features_and_labels_from_csv
from vehicle_behavior.features import parse_array_string, parse_coordinate_string

sns.set_theme(style="whitegrid")
```

```python
DATA_PATH = "../data/sample/sample_trajectories.csv"
df = extract_vehicle_labels(pd.read_csv(DATA_PATH))
features, labels = load_features_and_labels_from_csv(DATA_PATH)
print(f"Rows: {len(df)} | Valid trajectories: {len(features)}")
df["behavior_label"].value_counts()
```

```python
seq_lengths = [f["sequence_length"] for f in features]
plt.figure(figsize=(7, 4))
plt.hist(seq_lengths, bins=15, color="#3498db", edgecolor="white")
plt.title("Trajectory sequence lengths")
plt.xlabel("timesteps")
plt.ylabel("count")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4), sharey=True)
for ax, label in zip(axes, ["aggressive", "cooperative", "normal"]):
    row = df[df["behavior_label"] == label].iloc[0]
    speeds = parse_array_string(row["Speeds"])
    ax.plot(speeds, color="#e74c3c")
    ax.set_title(label)
    ax.set_xlabel("timestep")
plt.ylabel("speed")
plt.suptitle("Representative speed profiles")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, label in zip(axes, ["aggressive", "cooperative", "normal"]):
    row = df[df["behavior_label"] == label].iloc[0]
    coords = parse_coordinate_string(row["VehFrontCoords"])
    xs = [c[0] for c in coords]
    ys = [c[1] for c in coords]
    ax.plot(xs, ys, marker="o", markersize=2)
    ax.set_title(label)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
plt.suptitle("Representative position paths")
plt.tight_layout()
plt.show()
```
